```python
import os
```

```python
markedUpWords = []
```

```python
def getmarkedUpWords(file_path):
    with open(file_path) as f:
        flag = 0
        new_word = ""
        for line in f:
            for word in line.split():
                if word.startswith('<markup>'):
                    if "</markup>" in word or "<\markup>" in word:
                        new_word = word.replace("<markup>", "")
                        new_word = new_word.replace("<\markup>", "")
                        new_word = new_word.replace("</markup>", "")
                        markedUpWords.append(new_word)
                        print(new_word)
                        flag = 0
                    else:
                        flag = 1
                        new_word = word.replace("<markup>", "")
                        continue
                elif "</markup>" in word or "<\markup>" in word:
                    word = word.replace("</markup>", "")
                    word = word.replace("<\markup>", "")
                    new_word = new_word + " " + word
                    markedUpWords.append(new_word)
                    flag = 0
                    print(new_word)
                else:
                    if flag == 1:
                        new_word = new_word + " " + word
    return markedUpWords
```

```python
base_directory = os.path.dirname(os.getcwd())
data_directory = os.path.join(base_directory,"Dataset","Deepanshu")
file_list = os.listdir(data_directory)
marked_words=[]
count=0
for file in file_list:
    doc_id = file[0:3]  
    getmarkedUpWords(os.path.join(data_directory,file))
```
