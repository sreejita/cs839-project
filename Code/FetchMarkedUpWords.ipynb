```python
def getmarkedUpWords(file_path):
    with open(file_path) as f:
        flag = 0
        new_word = ""
        for line in f:
            for word in line.split():
                if word.startswith('<markup>'):
                    if "</markup>" in word or "<\markup>" in word:
                        new_word = word.replace("<markup>", "")
                        new_word = new_word.replace("<\markup>", "")
                        new_word = new_word.replace("</markup>", "")
                        markedUpWords.append(new_word)
                        print(new_word)
                        flag = 0
                    else:
                        flag = 1
                        new_word = word.replace("<markup>", "")
                        continue
                elif "</markup>" in word or "<\markup>" in word:
                    word = word.replace("</markup>", "")
                    word = word.replace("<\markup>", "")
                    new_word = new_word + " " + word
                    markedUpWords.append(new_word)
                    flag = 0
                    print(new_word)
                else:
                    if flag == 1:
                        new_word = new_word + " " + word
    return markedUpWords
```
